# Adding Element To List
    a) Append
    b) Insert
    c) extend

```python
#Append isn a bulid-in method used to add an item at the end of a list
languages = ['c','c#','java']
languages
```

```python
languages.append('python')
languages
```

```python
#Duplicate append() method to add multiple items
languages.append('ruby')
languages.append('javascript')
languages
```

```python
# A list as an item can be added
languages.append(['php','angular','react-native'])
languages
```

```python
# Insert() is a build-in method used to add an item at a specific position
# syntax : list.insert(position, value)
languages.insert(0,'Sql')
languages
```

```python
#Extend() mehtod is used to add all items of obe list in another list
#syntax : list1.extend(list2)
more_languages=['R','django','spring','vb-net']
languages.extend(more_languages)
languages
```

```python
# Changing an item, refer to the index of the item

list = ['rob','jennifer',1,2,'bob']
list
```

```python
list[2] = 'ida'
list
```

```python
# Changing multiple items
list[2:4]= ['jemmimah','paython','michael'] # [2:4[ means 4-1 => [2:3]
list
```

```python
# inser() method
# syntax : list.insert(index, item)

list.insert(1, 'ida')
list
```

# Removing List Item
    a) remove()
    b) pop()
    c) delete()
    d) Cancel()

```python
# Remove() remove the specified item
li=['john','mike','David']
li.remove('john')
li
```

```python
# pop(1) remove an item using the specified index

names=['jim','curtis','David']
names.pop(2)
names
```

```python
# del keyword remove an item a the specified index
del languages[4]
languages
```

```python
# to completed delete a list we use 'del'

del names
```

```python
# clear() used to empty a list

languages.clear()
languages
```

# Intro to List Comprehension

```python
#A list Comprehension provides a shorter syntax while creating a new list from the existing list name
# example select 3 names who start with J and adding then in the new list called  'j_names'

names = ['ahmed','julie','cynthia','james','jim']
j_names=[]
for name in names:
    if 'j' in name:
        j_names.append(name)
        
j_names


```

```python
# transfor the code above in list Comprehension

J_names2=[name for name in names if 'j' in name]
J_names2
```
